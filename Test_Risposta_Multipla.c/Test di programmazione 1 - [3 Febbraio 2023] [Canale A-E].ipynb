# Linker :
Il compito del linker è quello di riuscire a combinare i vari moduli oggetto file generati dal compilatore in un unico file eseguibile.
Ricordando come avviene la compilazione in un file C:
PreProcessing: il preprocessore elabora il codice sorgente gestendo direttive quali ad esempio librerie, macro...
Compiling: il compilatore (nel caso di c gcc) traduce il codice sorgente pre-processato in codice-assembly.
Assembling: l'assembler converte il codice assembly in codice oggetto, non ancora un eseguibile completo.
Linking(linker) : il linker prende uno o più file oggetto e li combina in un unico file eseguibile.

# La tecnica di ricerca binaria (o dicotomica) :
Teoria:
La ricerca binaria, o dicotomica, è un algoritmo di ricerca efficiente che viene utilizzato per trovare un elemento specifico all'interno di un insieme di dati ordinati (ad esempio, un array o una lista ordinata). A differenza della ricerca lineare, che controlla ogni elemento uno per uno, la ricerca binaria divide ripetutamente a metà la porzione di array che potrebbe contenere l'elemento cercato, riducendo drasticamente il numero di confronti necessari.
Esempio: Dizionario
